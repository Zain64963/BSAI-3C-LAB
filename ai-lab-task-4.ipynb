```python
import string

# Task 1: LUHN Algorithm
class Luhn:
    def check(self, card_number):
        digits = [int(d) for d in str(card_number)]
        check_digit = digits.pop()
        digits.reverse()

        for i in range(0, len(digits), 2):
            digits[i] *= 2
            if digits[i] > 9:
                digits[i] -= 9

        total = sum(digits) + check_digit
        return total % 10 == 0

    def run(self):
        card = input("Enter card number: ")
        if self.check(card):
            print("Valid Card Number")
        else:
            print("Invalid Card Number")


# Task 2: Remove Punctuations
class TextCleaner:
    def remove_punct(self, text):
        result = ""
        for char in text:
            if char not in string.punctuation:
                result += char
        return result

    def run(self):
        text = input("Enter a sentence: ")
        print("Without punctuation:", self.remove_punct(text))


# Task 3: Sort Sentence Alphabetically
class SentenceSorter:
    def sort_text(self, text):
        words = text.lower().split()
        words.sort()
        return " ".join(words)

    def run(self):
        text = input("Enter a sentence: ")
        print("Alphabetically sorted:", self.sort_text(text))


# Menu for Lab 4 Tasks
class Lab4:
    def menu(self):
        while True:
            print("\n--- LAB 4 TASK MENU ---")
            print("1. LUHN Algorithm")
            print("2. Remove Punctuations")
            print("3. Sort Sentence Alphabetically")
            print("4. Exit")

            choice = input("Choose an option (1-4): ")

            if choice == "1":
                luhn = Luhn()
                luhn.run()
            elif choice == "2":
                cleaner = TextCleaner()
                cleaner.run()
            elif choice == "3":
                sorter = SentenceSorter()
                sorter.run()
            elif choice == "4":
                print("Goodbye!")
                break
            else:
                print("Invalid choice, try again.")



lab = Lab4()
lab.menu()
```
